```python
import numpy as np
import scipy as sp
import scipy.stats as stats
import scipy.linalg as la
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
%matplotlib inline
```

# Three Coin Flip Problem

In the three coin flip problem, coin 1 with probablity $\lambda$ of heads is flipped. If heads, coin 2 with probablity $p$ of heads is flipped three times. Otherwise, coin 3 with probablity of $q$ is flipped 3 times. The number of heads of the three flips is observed and recorded, but the result of the first flip is kept hidden. The goal of the Expectation Maximization (EM) Algorithm will be to determine $\lambda$, $p$, and $q$

```python
#Code to generate the Coin Data
def generateCoinData(N,theta):
    """
    Accepts the number of data points to generate, and the parameters in the array
    theta
    """
    results = []
    for i in xrange(N):
        coin1 = (np.random.rand(1) <= theta[0])
        if coin1:
            results.append(np.random.rand(3) <= theta[1])
        else:
            results.append(np.random.rand(3) <= theta[2])
    return np.sum(np.array(results)*1,axis = 1)
```

The EM Algorithm to estimate $\lambda$, $p$, and $q$ is a two step process. 

In the Expectation Step (E Step) the currently estimated parameters $\lambda$, $p$, and $q$ are used compute $\mu$, the expected value of the log likelihood function of the parameters

In the Maximization Step (M Step) $\mu$ is then used to compute new $\lambda$, $p$, and $q$ parameters

```python
#Code to use the EM algorithm to estimate the parameters for the 3 coin toss problem
def em_coinToss(data,lam,p,q,N):
    """
    Accepts the data, initial guesses for the parameters and number of iterations
    """
    thetas = np.zeros((N,3))
    thetas[0] = (lam,p,q)
    for i in xrange(1,N):
        #E Step
        mu = (lam *p**data *(1 - p)**(3 - data)/
              (lam *p**data *(1 - p)**(3 - data) +
               (1 - lam)*q**data * (1 - q)**(3 - data)))
        
        #M Step
        lam = 1./data.size * np.sum(mu)
        p = 1./np.sum(mu) *np.sum(data*mu/3.)
        q = 1./np.sum(1.-mu) *np.sum(data*(1.-mu)/3.)
        thetas[i] = (lam,p,q)
    return lam,p,q,thetas
```

Now we test the EM Algorithm for the three coin problem

```python
theta = [.4,.9,.2]
guess = [.3,.4,.1]
samples = 100
iters = 100
data = generateCoinData(samples,theta)
lam,p,q,thetas = em_coinToss(data,*guess,N=iters)
print "Actual ",theta
print "Estimated", [lam,p,q]
```

With as few as 100 samples, the algorithm estimated the parameters reasonably closely.
We can plot the convergence of the algorithm

```python
I = np.arange(iters)
plt.plot(I,thetas[:,0],"b-",label=r"Estimated $\lambda$")
plt.plot(I,thetas[:,1],"g-",label=r"Estimated $p$")
plt.plot(I,thetas[:,2],"r-",label=r"Estimated $q$")
plt.plot(I,theta[0]*np.ones(iters),"b--")
plt.plot(I,theta[1]*np.ones(iters),"g--",)
plt.plot(I,theta[2]*np.ones(iters),"r--")
plt.legend()
plt.show()
```

However, the algorithm performs best with larger number of data points

```python
samples = np.arange(100,2000,200)
m = len(samples)
data = generateCoinData(samples[-1],theta)
thetas = np.zeros((m,3))
iters = 100

theta = [.4,.9,.2]
guess = [.7,.4,.3]
for i,s in enumerate(samples):
    lam,p,q,guesses = em_coinToss(data[:s],*guess,N=iters)
    thetas[i] = (lam,p,q)


I = np.arange(m)
plt.plot(I,thetas[:,0],"b-",label=r"Estimated $\lambda$")
plt.plot(I,thetas[:,1],"g-",label=r"Estimated $p$")
plt.plot(I,thetas[:,2],"r-",label=r"Estimated $q$")
plt.plot(I,theta[0]*np.ones(m),"b--")
plt.plot(I,theta[1]*np.ones(m),"g--",)
plt.plot(I,theta[2]*np.ones(m),"r--")
plt.legend(loc=0)
plt.show()
```

# Gaussian Mixture

In the Gaussian Mixture problem, an $n$ sided dice is rolled to determine a gaussian to sample from, each side having a probablity of $\lambda_j$. That gaussian, with mean $\mathbf{c}_j$ and standard deviation $\mathbf{\sigma}_j$, is sampled $d$ times. Like in the coin flip problem, the sample from the gaussian is observed but the outcome of the dice roll is hidden. The goal of the problem is to determine the parameters $\lambda_j$,$\mathbf{c}_j$, and $\\mathbf{sigma}_j$

```python
#Code to generate Gaussian Mixture data
def generateGaussianData(N,lam,c,sig):
    data = []
    cumsum_lam = np.cumsum(lam)
    for i in xrange(0,N):
        d = np.random.rand(1)
        j = len(np.where(d > cumsum_lam)[0])
        
        data.append(np.random.normal(loc=c[j],scale=sig[j]))
    return np.array(data)
```

The EM Algorithm for Gaussian Mixture is similar to the coin flip problem

In the Expectation Step (E Step) the currently estimated parameters $\lambda_j$, $\mathbf{c}_j$, and  $\mathbf{sigma}_j$ are used compute $\mathbf{w}_j^i$, weights for each observation

In the Maximization Step (M Step) $\mathbf{w}_j^i$ is then used to compute new $\lambda_j$, $\mathbf{c}_j$, and $\mathbf{sigma}_j$  parameters

The algorithm is a bit more complex due to the extra dimensionality

```python
def gaussianMixture(x,n,N):
    
    if(len(x.shape) == 1):
        x = x.reshape(x.size,1)
    d = x.shape[1]
    
    lam = np.random.rand(n)
    lam /= np.sum(lam)
    
    c = np.random.rand(n,d)
    sig = np.random.rand(n,d)
    
    
    def pdf(x,c,sig):
        return (1./((2.*np.pi)**(d/2.)*sig**d) *
                np.exp( -la.norm(x-c)**2/(2*sig**2.)))
    for i in xrange(N):
        
        w = np.array([np.array([lam1*pdf(x1,c1,sig1) for lam1,c1,sig1 in zip(lam,c,sig)])
                      for  x1 in x])
        w /= w.sum(axis=1).reshape((w.shape[0],1,w.shape[2]))
        
        lam = np.mean(w,axis=0).mean(axis=1)
        c = (1./w.sum(axis=0)*np.sum(
                w*x.reshape(x.shape[0],1,x.shape[1]),axis=0))
        
        xmc = np.linalg.norm(x.reshape((x.shape[0],1,x.shape[1]))-
                        c.reshape((1,c.shape[0],c.shape[1])),axis=2)**2.0
        sig = np.sqrt(1./(w.sum(axis=0)*d)*
                      np.sum(w * xmc.reshape((xmc.shape[0],xmc.shape[1],1)),
                             axis=0))
        
    return lam,c,sig
```

We now test the algorithm in 1 d

```python
theta_lam = np.array( [.1, .6, .15, .15])
theta_c = np.array([.1, .9, .5, .3])
theta_sig = np.array([.2, .1, .05, .1])
data = generateGaussianData(2000,theta_lam,theta_c,theta_sig)
lam,c,sig = gaussianMixture(data,4,300)
```

```python
#Helper function to plot the data with Gaussians
def plotGaussians(lams,cs,sigs,N,data): 
    fig, ax1 = plt.subplots()
    ax1.hist(data,50,alpha=.1)
    
    ax2 = ax1.twinx()
    x = np.linspace(np.min(cs)-np.max(sigs)*4,np.max(cs)+np.max(sigs)*4,N)
    totalPDF = np.zeros(N)
    for i,(lam,c,sig) in enumerate(zip(lams,cs,sigs)):
        norm = stats.norm(loc=c,scale=sig)
        ax2.plot(x,lam*norm.pdf(x),label="PDF {}".format(i+1),linewidth=2)
        totalPDF += lam*norm.pdf(x)
    ax2.plot(x,totalPDF,label="Total",linewidth=2)
    plt.legend(loc=2)
```

```python
plt.hist(data,50)
plt.title("Histogram of Data")
plt.show()

plotGaussians(theta_lam,theta_c,theta_sig,500,data)
plt.title("Actual Gaussians")
plt.show()

plotGaussians(lam,c.ravel(),sig.ravel(),500,data)
plt.title("Estimated Gaussians")
plt.show()
```

The EM algorithm also works well in 2d, as long as the Gaussians are fairly separated

```python
np.random.seed(100)
theta_lam = np.array( [.1, .6, .15, .15])
theta_c = np.array([[.1,.1], [.9,.8], [.5,.1], [.3,.7]])
theta_sig = np.array([[.02,.02],[.015,.03], [.05,.03],[.07,0.08]])
data = generateGaussianData(500,theta_lam,theta_c,theta_sig)
lam,c,sig = gaussianMixture(data,4,100)
```

```python
#Helper function for plotting the data with gaussian ellispses
from matplotlib.patches import Ellipse
def plot2dGaussians(lams,cs,sigs,data):
    plt.figure()
    ax = plt.gca()
    
    plt.scatter(data[:,0],data[:,1])
    for i,(lam,c,sig) in enumerate(zip(lams,cs,sigs)):
        e = Ellipse(xy=(c),width=sig[0]*4,height=sig[1]*4)
        e.set_alpha(.3)
        e.set_color('r')
        ax.add_patch(e)
```

```python
plot2dGaussians(theta_lam,theta_c,theta_sig,data)
plt.title("Actual Gaussians")
plt.show()

plot2dGaussians(lam,c,sig,data)
plt.title("Estimated Gaussians")
plt.show()
```

The algorithm does very well for these spread out gaussians
