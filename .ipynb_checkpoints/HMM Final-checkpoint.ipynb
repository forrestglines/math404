# Three Coin Problem

The three coin problem is stated as follows:

A man behind a curtain flips a coin (not necessarily fair) that you are unable to observe. Assume this initial coin has some unknown probability $\lambda$ of being heads. If the initial coin is heads, coin 1 is flipped three times, where coin 1 has an unknown probability of being heads of $p$. However, if the initial coin is tails, then coin 2 is flipped three times, where coin 2 has an unknown probability of being heads of $q$. The results of the three coin flips are reported to you-- however whether they were generated by coin 1 or coin 2 is unknown. The problem is to determine the underlying parameters $\lambda$, $p$, and $q$.

The following code snippet generates the data:

```python
#Generates coinflip data as a Hidden Markov Model
#For each of n samples, a coin X with lam probability of coming up heads is flipped
#If X is heads, coin 1 is flipped nn times, where coin 1 has a probability of heads of p1
#If X is tails, coin 2 is flipped nn times, where coin 2 has a probability of heads of p2
#n is the number of samples
#nn is the number of coin flips in each sample
def generate_coinflip_data(n, nn, lam, p1, p2):
    def flip_1(m):
        res = np.random.binomial(m, p1)
        return res
    
    def flip_2(m):
        res = np.random.binomial(m, p2)
        return res
    
    data = []
    for i in xrange(n):
        init_flip = random.random(1)
        if init_flip < lam:
            data.append(flip_1(nn))
        else:
            data.append(flip_2(nn))
            
    return np.array(data)
```

One common solution to this problem is to use the Expectation Maximization algorithm (EM) to estimate the unknown parameters $\lambda$, $p$, and $q$. The EM method is an iterative method with two steps that converges to a local maximum for the parameters. The first step is the "E" step ("Expectation" step), which uses the current estimates of the parameters to generate a log-likelihood function. The second step is the "M" step ("Maximization" step), which uses maximum likelihood estimation (MLE) to maximize the function found in the "E" step. The MLE estimates that maximize the likelihood function are used as the new parameters in the next iteration of the "E" and "M" steps. This algorithm eventually converges to a local maximum.

The EM_alg() function below is a working example of the EM algorithm for the Three Coin problem. Given data and an initial guess, the algorithm estimates the unknown coin parameters.

```python
#Attempts to back out the probabilities of lam, p1, and p2 from the generating function above given data
#Requires an initial guess of l_0, p1_0, and p2_0
#This algorithm only finds local maximums, and sometimes is confused by the data if poor initial guesses are given
def EM_alg(data, j, l_0 = .4, p1_0 = .3, p2_0 = .6, tol=.0001):
    iters = 0
    diff = 1
    lam = l_0
    p1 = p1_0
    p2 = p2_0
    j = float(j)
    
    while np.abs(diff) > tol:
        iters += 1
        
        h = data
        l = lam
        p = p1
        q = p2
        mu = (l * p ** h * (1 - p) ** (j - h))/((l * p ** h * (1 - p) ** (j - h)) + ((1 - l) * q ** h * (1 - q)**(j - h)))
                
        lam_new = np.sum(mu/(1.0*len(data)))
        
        p_new = np.sum((h / j) * mu)/float(np.sum(mu))
       
        q_new = np.sum((h / j) * (1 - mu)) / float(np.sum(1 - mu))
        
        p1 = p_new
        p2 = q_new
        
        diff = lam - lam_new
        lam = lam_new
        
    return lam, p1, p2
```

Now, we will test the algorithm on the Three Coin problem:

```python
actual_parameters = [.3, .8, .2]
samples = 100
flips_per_sample = 3
d = generate_coinflip_data(samples, flips_per_sample, actual_parameters[0], actual_parameters[1], actual_parameters[2])
res = EM_alg(d, flips_per_sample, l_0 = 0.5, p1_0 = .4, p2_0 = .1, tol=.0000001)
print "Actual parameter values:", "lambda =", actual_parameters[0], ", p =", actual_parameters[1], ", q =", actual_parameters[2]
print "Estimated parameter values:", "lambda =", res[0], ", p =", res[1] , ", q =", res[2]
```

The estimates are quite accurate, even for a small number of samples. However, when p and q are similar, it is difficult to determine p and q, as in the following example. Here, 5 separate tests are performed on 5 sets of sample data generated by the same process. The estimates wildly differ:

```python
actual_parameters = [.3, .6, .7]
samples = 100
flips_per_sample = 3
print "Actual parameter values:", "lambda =", actual_parameters[0], ", p =", actual_parameters[1], ", q =", actual_parameters[2]
for i in xrange(5):
    d = generate_coinflip_data(samples, flips_per_sample, actual_parameters[0], actual_parameters[1], actual_parameters[2])
    res = EM_alg(d, flips_per_sample, l_0 = 0.5, p1_0 = .7, p2_0 = .5, tol=.0000001)
    print "Trial " + str(i+1) + " estimated parameter values:", "lambda =", res[0], ", p =", res[1] , ", q =", res[2]
```

One possible solution is simply to increase the data sample size. Below, we increase the number of data points by an order of magnitude (100 to 1000 samples of 3 flips each). However, it is very slow and the estimates still are exceptionally poor:

```python
actual_parameters = [.3, .6, .7]
samples = 1000
flips_per_sample = 3
print "Actual parameter values:", "lambda =", actual_parameters[0], ", p =", actual_parameters[1], ", q =", actual_parameters[2]
for i in xrange(5):
    d = generate_coinflip_data(samples, flips_per_sample, actual_parameters[0], actual_parameters[1], actual_parameters[2])
    res = EM_alg(d, flips_per_sample, l_0 = 0.5, p1_0 = .7, p2_0 = .5, tol=.0000001)
    print "Trial " + str(i+1) + " estimated parameter values:", "lambda =", res[0], ", p =", res[1] , ", q =", res[2]
```

Additionally, if the number of coin flips in each sample is increased, it becomes much easier to estimate $p$ and $q$. This is because with only three flips per sample, it is hard to determine the probability of heads accurately for the coin with which it was generated. In this case, the similarity of $p$ and $q$ complicate this difficulty. With more draws per sample, the EM algorithm converges much more accurately, as shown below where the number of flips per sample is increased from 3 to 10:

```python
actual_parameters = [.3, .6, .7]
samples = 100
flips_per_sample = 20
print "Actual parameter values:", "lambda =", actual_parameters[0], ", p =", actual_parameters[1], ", q =", actual_parameters[2]
for i in xrange(5):
    d = generate_coinflip_data(samples, flips_per_sample, actual_parameters[0], actual_parameters[1], actual_parameters[2])
    #print len(np.where(np.sum(d, axis=1)/20.0 > .6)[0])/1000.
    res = EM_alg(d, flips_per_sample, l_0 = 0.3, p1_0 = .55, p2_0 = .75, tol=.0000001)
    print "Trial " + str(i+1) + " estimated parameter values:", "lambda =", res[0], ", p =", res[1] , ", q =", res[2]
```

Overall, the EM algorithm performs well on this problem, except for very tricky cases, such as with $p$ and $q$ set at $.6$ and $.7$.

#Gaussian Mixture Model

The Gaussian Mixture Model problem is similar to the Three Coin problem, in that again, the generating process has a hidden parameter. The problem can be described as follows: $n$ normal distributions are specified and numbered 1 through $n$, each with some parameters $\mu_i$ and $\sigma_i$ for the mean and standard deviation. A dice with $n$ sides (that is possibly unfair) is rolled behind a curtain. The roll of the dice, $m$, determines which normal distribution will be sampled. The only observation given to the viewer is a draw from the normal distribution with parameters $\mu_m$ and $\sigma_m$, however these parameters (and $m$) are unknown to the observer. The problem is, like the coin problem, to estimate the parameters $\lambda_i$ for $i \in [1, ..., n]$ (the probability of the dice rolling $i$), $\mu_1$, ..., $\mu_n$, and $\sigma_1$, ..., $\sigma_n$.

The following code generates the GMM data. For simplification purposes, the code only generates univariate data:

```python
def generate_mixture_model_data(n, lam, means, sigs):
    data = []
    iters = 0
    while iters < n:
        iters +=1
        draw = random.random(1)
        agg_lam = np.array([np.sum(lam[:i]) for i in xrange(len(lam))])
        i = len(np.where(draw > agg_lam)[0]) - 1
        data.append(random.normal(loc=means[i], scale=sigs[i]))
    return np.array(data)
```

Similar to the Three Coin problem above, this solver code below uses the EM algorithm. It is considerably more complex due to the increase in number of parameters and the increase in complexity of the log-likelihood. However, it works in much the same way. Note that it "intelligently" chooses the starting values for the $\mu_i$ and $\sigma_i$ parameters, while the $\lambda_i$ initial values are guessed randomly.

```python
def EM_GMM_alg(data, n, tol=.001, verbose=0, robust=False):
    iters = 0
    diff = 1
    l_0 = np.random.rand(n)
    mean_max = np.max(data)
    mean_min= np.min(data)
    
    m_0 = np.linspace(mean_min, mean_max, n)
    s_0 = 0.15 * np.ones(n)
    l_0 /= np.sum(l_0)
    lam = l_0
    m = m_0
    s = s_0
    
    def pdf(x, m, s):
        d = 1.0
        return (1/((2.0 * np.pi)**(d/2.0) * s**d))*np.exp((-np.linalg.norm(x-m)**2)/(2.0*s**2))
    
    while np.abs(diff) > tol:
        if verbose == 2:
            print "\rLam:" + str(lam) +" Means:" + str(m) + " Sigs:" + str(s),
        elif verbose == 1:
             print "\rIters: " + str(iters),
        p = np.array([[pdf(data[i], m[j], s[j]) for j in xrange(len(m))] for i in xrange(len(data))])
        lam_old = lam.copy()
        p *= lam
        z_s = np.sum(p, axis=1)
        w = p/z_s.reshape((len(z_s), 1))
        lam = np.mean(w, axis=0)
        w_s = np.sum(w, axis=0)
        w_d = w*data.reshape((len(data), 1))
        w_d_s = np.sum(w_d, axis=0)
        c = w_d_s/w_s
        to_norm = np.array([data - c[j] for j in xrange(len(m))])
        to_norm = np.swapaxes(to_norm, 0, 1)
        ss = np.sum(w*(to_norm**2.0), axis=0)
        s = np.sqrt(ss/(1 * w_s))
        m = c
        diff = np.linalg.norm(lam - lam_old)
        if robust:
            diff = [np.linalg.norm(lam - lam_old), np.linalg.norm(c - old_c), np.linalg.norm(sig - old_sig)]
            diff = np.max(diff)
        iters += 1
        
    #ll = np.sum([[np.log(lam[j]*pdf(data[i], m[j], s[j])) for j in xrange(len(lam))] for i in xrange(len(data))])
    return lam, m, s
```

Additionally, along with the solver, a plotting function is used that nicely displays the estimated versus actual probability density functions for the actual and estimated GMMs:

```python
def plot_actual_vs_estimated(d, actual_lams, actual_means, actual_sigs, estimated_lams, estimated_means, estimated_sigs):
    #Plot actual
    avg_mean = np.mean(actual_means)
    avg_sig = np.mean(actual_sigs)
    PLOTTING_DEV = 6.0
    x = np.linspace(avg_mean - PLOTTING_DEV * avg_sig, avg_mean + avg_sig * PLOTTING_DEV, 500)
    
    cumulative_norm = np.zeros_like(x)
    
    plt.subplot(3, 1, 1)
    plt.hist(d, bins=100)
    plt.title("Histogram of Data")
    plt.show()

    plt.subplot(3, 1, 2)
    for i in xrange(len(actual_lams)):
        norm = scipy.stats.norm(loc=actual_means[i], scale=actual_sigs[i])
        plt.plot(x, norm.pdf(x)*actual_lams[i], label="PDF " + str(i))
        cumulative_norm += actual_lams[i]*norm.pdf(x)
        
    plt.plot(x, cumulative_norm, label="Cumulative PDF")
    plt.title("Actual Gaussian Distributions")
    plt.legend(loc="upper left")
    
    cumulative_norm_estimate= np.zeros_like(x)
    
    plt.subplot(3, 1, 3)
    for i in xrange(len(estimated_lams)):
        norm = scipy.stats.norm(loc=estimated_means[i], scale=estimated_sigs[i])
        plt.plot(x, norm.pdf(x)*estimated_lams[i], label="PDF " + str(i))
        cumulative_norm_estimate += estimated_lams[i]*norm.pdf(x)
        
    plt.plot(x, cumulative_norm_estimate, label="Cumulative PDF")
    plt.legend(loc="upper left")
    plt.title("Estimated Gaussian Distributions")
    plt.show()
    
    plt.plot(x, cumulative_norm, label="Actual")
    plt.plot(x, cumulative_norm_estimate, label="Estimate")
    plt.title("Actual vs. Estimated Total Distributions")
    plt.legend(loc="upper left")
    plt.show()
```
